```python
loaded_model = YOLO("yolov8n_road.pt")
```

```python
import torch
# Load and preprocess image
image = cv2.imread("Traffic-signals.png")

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
image = cv2.resize(image, (640, 640))
image = image/255
# Convert to tensor and add batch dimension
image = torch.from_numpy(image).to(loaded_model.device)
image = torch._cast_Float(image)
image = image.unsqueeze(0)  # Shape: [1, 3, 640, 640]

# Run model inference (raw tensor output)
with torch.no_grad():
    raw_tensor = loaded_model.model(image)  # Access the underlying PyTorch model

# raw_tensor is a list of tensors (e.g., for detection: [1, 84, 8400])
print(raw_tensor[0].shape)  # Example output: torch.Size([1, 84, 8400])
```
