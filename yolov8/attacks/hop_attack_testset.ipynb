```python
import sys
import io
class SuppressPrints:  # Context manager to silence function output
    def __enter__(self):
        self._original_stdout = sys.stdout
        sys.stdout = io.StringIO()
    def __exit__(self, exc_type, exc_val, exc_tb):
        sys.stdout = self._original_stdout
```

```python
import os
import torch
from torchvision.utils import save_image
from pathlib import Path


# Configure paths
test_images_dir = "/content/test/images/"
output_dir = "/content/adversarial_test_images_hop/"
os.makedirs(output_dir, exist_ok=True)

# Get test images
test_images = [f for f in Path(test_images_dir).glob('*') if f.suffix in ('.jpg', '.png', '.jpeg')]

i=0
# Attack loop with PNG saving
for img_path in test_images:
    x_orig = preprocess_image(str(img_path))
    with SuppressPrints():  # Silence function internals
        adv_tensor = hop_skip_jump_attack(model,
                                          x_orig,
                                          epsilon=0.1, delta=0.1,
                                          batch_size=100,
                                          norm="l2",
                                          max_queries=30,
                                          max_iters=10)
    # Save as PNG with proper quantization
    output_path = os.path.join(output_dir, f"{img_path.stem}.png")

    # Clone, detach, and save
    if adv_tensor is None:
      continue

    save_image(
        adv_tensor.clone().detach().cpu(),  # Move to CPU if needed
        output_path,
        normalize=True  # Automatically scales input to [0, 255] if in [0, 1]
    )
    i= i+1
    if i % 10 == 0:
      print("------------------------")
      print("")
      print("")
      print("")
      print(f"image number {i}")
      print("")
      print("")
      print("")
      print("------------------------")
```

```python

import matplotlib.pyplot as plt
import random
import os

# Assuming images are in /content/adversarial_dataset/images
image_dir = "/content/adversarial_dataset/images"

# Get a list of image files
image_files = [f for f in os.listdir(image_dir) if os.path.isfile(os.path.join(image_dir, f))]

# Select 3 random images
num_samples = 4
random_images = random.sample(image_files, min(num_samples, len(image_files)))

# Display the images
plt.figure(figsize=(12, 4))
for i, image_file in enumerate(random_images):
    img_path = os.path.join(image_dir, image_file)
    try:
        img = plt.imread(img_path)
        plt.subplot(1, num_samples, i + 1)
        plt.imshow(img)
        plt.axis('off')
    except Exception as e:
        print(f"Error displaying {image_file}: {e}")
plt.show()
```
